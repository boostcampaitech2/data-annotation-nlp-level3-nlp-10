```python
import os
dir_path = './rawdata'
flist = os.listdir(dir_path)
print(flist)
```

```python
from kss import split_sentences

splited = []
for fname in flist:
    f = open(os.path.join(dir_path, fname), 'r', encoding='utf-8')
    text = f.read()
    # print(text)
    splited.append(split_sentences(text))
    # while True:
    #     line = f.readline()
    #     if not line: break
    #     text.append(line)

    f.close()
```

Mecab으로 pos tagging후 적어도 주어, 동사가 있는 경우만 문장으로 추출할 계획

```python
from konlpy.tag import Mecab
```

```python
mecab = Mecab()
test = []
rej = []
for text in splited:
    for line in text:
        res = mecab.pos(line)
        if line == '같이 보기':
            rej.append(line)
            continue
        flag0 = False; flag1 = False
        for tup in res:
            if ('N') in tup[1]: flag0 = True
            if 'V' in tup[1] or 'E' in tup[1]: flag1 = True
        if flag0 and flag1: test.append(line)
        else: rej.append(line)
```
